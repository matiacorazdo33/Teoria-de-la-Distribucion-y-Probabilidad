```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns

# Tarea 3 (Modificada): Variable Aleatoria Discreta - Distribución Binomial
# Caso: X = número de paquetes exitosos en un envío de n=20 con p=0.9
# Matias Labanda

# 1. Parámetros de la distribución binomial
n = 20  # número de ensayos (paquetes)
p = 0.9 # probabilidad de éxito

# 2. Definir la distribución y rango de valores
dist_binomial = binom(n, p)
x_valores = np.arange(0, n + 1)

# 3. Calcular PMF y CDF
pmf_binom = dist_binomial.pmf(x_valores)
cdf_binom = dist_binomial.cdf(x_valores)

# 4. Visualización (Siguiendo el formato de la Tarea 3 original)
plt.figure(figsize=(12, 5))
sns.set_style("whitegrid")

# Subplot 1: PMF
plt.subplot(1, 2, 1)
plt.stem(x_valores, pmf_binom, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.fill_between(x_valores, pmf_binom, alpha=0.2, color='blue')
plt.xlabel('x (Paquetes exitosos)')
plt.ylabel('p(x) = P(X = x)')
plt.title('PMF: Distribución Binomial (Control de Calidad)')
plt.grid(alpha=0.3)

# Subplot 2: CDF
plt.subplot(1, 2, 2)
plt.step(x_valores, cdf_binom, where='post', color='red', linewidth=2)
plt.scatter(x_valores, cdf_binom, color='red', zorder=5)
plt.xlabel('x')
plt.ylabel('F(x) = P(X <= x)')
plt.title('CDF: Distribución Binomial Acumulada')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 5. Cálculo de probabilidades específicas
prob_18 = dist_binomial.pmf(18)
prob_menor_15 = dist_binomial.cdf(15)
prob_entre_17_19 = dist_binomial.cdf(19) - dist_binomial.cdf(16)

print(f"Resultados de Probabilidad:")
print("-" * 30)
print(f"a) P(X = 18) = {prob_18:.4f}")
print(f"b) P(X <= 15) = {prob_menor_15:.4f}")
print(f"c) P(17 <= X <= 19) = {prob_entre_17_19:.4f}")
```
